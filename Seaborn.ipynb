
# plotting using seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
import pandas as pd
```

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", style="time", size="size")
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", kind="scatter", hue="sex", style="time", size="size")
```

```python
iris.head()
```

```python
sns.pairplot(iris, hue="species")
```

```python
g=sns.PairGrid(data=iris)
g.map(sns.scatterplot)
```

```python
g=sns.PairGrid(data=iris, hue="species")
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
g=sns.PairGrid(data=iris, hue="species", vars=["sepal_width", "petal_width"])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
sns.jointplot(data=iris, x="sepal_width", y="petal_width")
```

```python
sns.jointplot(data=iris, x="sepal_width", y="petal_width", kind="kde")
```

```python
g=sns.JointGrid(data=iris, x="sepal_width", y="petal_width")
g.plot(sns.boxplot, sns.kdeplot)
```
