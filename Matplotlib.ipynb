```python
df=pd.read_csv("batter.csv")
df.shape
```

```python
df=df.sample(25, random_state=5)
df.head()
```

```python
plt.scatter(df["avg"], df["strike_rate"])
```

```python
plt.figure(figsize=(19,10))

plt.scatter(df["avg"], df["strike_rate"], s=df["runs"])

for i in range(df.shape[0]):
    plt.text(df["avg"].values[i], df["strike_rate"].values[i], df["batter"].values[i])
    
plt.show()
```

```python
plt.figure(figsize=(19,10))

plt.scatter(df["avg"], df["strike_rate"], s=df["runs"])

plt.axhline(130)
plt.axvline(30)


for i in range(df.shape[0]):
    plt.text(df["avg"].values[i], df["strike_rate"].values[i], df["batter"].values[i])
    
plt.show()
```

```python
df=pd.read_csv("batter.csv")
df.head()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1, sharex=True)

ax[0].scatter(df["avg"], df["runs"])
ax[0].set_title("avg vs runs")
ax[0].set_ylabel("runs")

ax[1].scatter(df["avg"], df["strike_rate"] )
ax[1].set_title("avg vs strike rate")
ax[1].set_ylabel("strike rate")
ax[1].set_xlabel("avg")

plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)

ax[0,0].scatter(df["avg"], df["runs"])
ax[0,1].scatter(df["avg"], df["strike_rate"])
ax[1,0].hist(df["avg"])
ax[1,1].hist(df["strike_rate"])

plt.show()
```

```python
fig=plt.figure()

ax1=fig.add_subplot(2,2,1)
ax1.scatter(df["avg"], df["runs"])

ax2=fig.add_subplot(2,2,2)
ax2.scatter(df["avg"], df["strike_rate"])
                
ax3=fig.add_subplot(2,2,3)
ax3.hist(df["avg"])

ax4=fig.add_subplot(2,2,4)
ax4.hist(df["strike_rate"])

plt.show()
```

```python
fig=plt.figure()
ax=plt.subplot(projection="3d")
ax.scatter3D(df["runs"], df["avg"], df["strike_rate"])

ax.set_title("batsman analysis")
ax.set_xlabel("runs")
ax.set_ylabel("avg")
ax.set_zlabel("strike_rate")

plt.show()
```

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
```

```python
z=xx**2+yy**2
z.shape
```

```python
ax=plt.subplot(projection="3d")
f=ax.plot_surface(xx, yy, z, cmap="viridis")
fig.colorbar(f)
plt.show()
```

```python
z=np.sin(xx)+np.tan(yy)

ax=plt.subplot(projection="3d")
f=ax.plot_surface(xx, yy, z, cmap="viridis")
fig.colorbar(f)
plt.show()
```

```python
z=np.sin(xx)+np.cos(yy)

ax=plt.subplot(projection="3d")
f=ax.plot_surface(xx, yy, z, cmap="viridis")
fig.colorbar(f)
plt.show()
```

```python
z=np.log(xx)+np.log(yy)

ax=plt.subplot(projection="3d")
f=ax.plot_surface(xx, yy, z, cmap="viridis")
fig.colorbar(f)
plt.show()
```

```python
ax=plt.subplot()
f=ax.contour(xx, yy, z, cmap="viridis")
fig.colorbar(f)
plt.show()
```

```python
ax=plt.subplot()
f=ax.contourf(xx, yy, z, cmap="viridis")
fig.colorbar(f)
plt.show()
```

```python
z=np.sin(xx)+np.cos(yy)

ax=plt.subplot()
f=ax.contourf(xx, yy, z, cmap="viridis")
fig.colorbar(f)
plt.show()
```
